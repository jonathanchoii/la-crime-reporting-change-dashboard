```python
# Checking full row counts and data coverage
import requests

# Shows 1. Total published rows 2. Earliest occurrence date 3. Latest occurrence date
dataset_ids = {
    "legacy_2020_2024": "2nrs-mtv8",
    "nibrs_2024_2025": "y8y3-fqfu",
    "nibrs_2026_present": "k7nn-b2ep",
}

coverage_results = []

for dataset_name, dataset_id in dataset_ids.items():
    url = f"https://data.lacity.org/resource/{dataset_id}.json"

    parameters = {
        "$select": (
            "count(*) as row_count, "
            "min(date_occ) as min_date_occ, "
            "max(date_occ) as max_date_occ"
        )
    }

    response = requests.get(url, params=parameters, timeout=60)
    response.raise_for_status()

    result = response.json()[0]
    result["dataset_name"] = dataset_name
    coverage_results.append(result)

coverage = pd.DataFrame(coverage_results)

coverage["row_count"] = coverage["row_count"].astype(int)

coverage = coverage[
    ["dataset_name", "row_count", "min_date_occ", "max_date_occ"]
]

display(coverage)
```

```python
# Compare occurrence dates with report dates
reporting_coverage_results = []

for dataset_name, dataset_id in dataset_ids.items():
    url = f"https://data.lacity.org/resource/{dataset_id}.json"

    parameters = {
        "$select": (
            "count(*) as row_count, "
            "min(date_rptd) as min_date_rptd, "
            "max(date_rptd) as max_date_rptd, "
            "min(date_occ) as min_date_occ, "
            "max(date_occ) as max_date_occ"
        )
    }

    response = requests.get(url, params=parameters, timeout=60)
    response.raise_for_status()

    result = response.json()[0]
    result["dataset_name"] = dataset_name
    reporting_coverage_results.append(result)

reporting_coverage = pd.DataFrame(reporting_coverage_results)

reporting_coverage["row_count"] = (
    reporting_coverage["row_count"].astype(int)
)

reporting_coverage = reporting_coverage[
    [
        "dataset_name",
        "row_count",
        "min_date_rptd",
        "max_date_rptd",
        "min_date_occ",
        "max_date_occ",
    ]
]

display(reporting_coverage)
```

```python
# Downloading the complete raw files
from pathlib import Path

current_directory = Path.cwd()

if (current_directory / "README.md").exists():
    project_root = current_directory
else:
    project_root = current_directory.parent

raw_data_directory = project_root / "data" / "raw"
raw_data_directory.mkdir(parents=True, exist_ok=True)

print("Project root:", project_root)
print("Raw data directory:", raw_data_directory)
```

```python
download_urls = {
    "lapd_legacy_2020_2024.csv": (
        "https://data.lacity.org/api/views/2nrs-mtv8/"
        "rows.csv?accessType=DOWNLOAD"
    ),
    "lapd_nibrs_2024_2025.csv": (
        "https://data.lacity.org/api/views/y8y3-fqfu/"
        "rows.csv?accessType=DOWNLOAD"
    ),
    "lapd_nibrs_2026_present.csv": (
        "https://data.lacity.org/api/views/k7nn-b2ep/"
        "rows.csv?accessType=DOWNLOAD"
    ),
}
```

```python
for filename, url in download_urls.items():
    destination = raw_data_directory / filename

    if destination.exists():
        print(f"{filename} already exists. Skip download.")
        continue

    print(f"Downloading {filename}...")

    with requests.get(
        url,
        stream=True,
        timeout=(30, 300)
    ) as response:
        response.raise_for_status()

        with open(destination, "wb") as file:
            for chunk in response.iter_content(chunk_size=1024 * 1024):
                if chunk:
                    file.write(chunk)

    size_mb = destination.stat().st_size / (1024 ** 2)
    print(f"Saved {filename}: {size_mb:.2f} MB")
```
